```python
### Welcome Screen and Menu to begin
while True:
    print("Hello! I'm A.V.A., your Academic Virtual Assistant")
    print("")
    print("Please select an option from the menu:")
    print("")
    print("    1 - Assignment Tracker/ToDo List") 
    print("    2 - Final Grade Calculator")
    print("    3 - Get an Affirmation")
    print("    4 - Exit")
##should add a loop somewhere...
#Stores user selection for the if statement
    function = int(input("Enter a number from the menu aboove:"))

#nested if that goes into each function
#Assignment Tracker
    if function==1:
        print("You have chosen to make an Assignemnt Tracker/ToDo List")
        assignments = []

        def show_menu():
            print("\nüìò Assignment Tracker Menu")
            print("1 - Add an Assignment")
            print("2 - View All Assignments")
            print("3 - Mark an Assignment as Complete")
            print("4 - Delete an Assignment")
            print("5 - Exit to Main Menu")

        while True:
            show_menu()
            choice = input("Choose an option (1-5): ")

            if choice == "1":
                assignment = input("Enter the assignment name: ")
                due_date = input("Enter the due date (optional): ")
                assignments.append({"task": assignment, "due": due_date, "completed": False})
                print(f"‚úÖ '{assignment}' has been added to your list!")

            elif choice == "2":
                if not assignments:
                    print("üì≠ No assignments in your list yet.")
                else:
                    print("\nYour Assignments:")
                    for i, a in enumerate(assignments, start=1):
                        status = "‚úÖ Done" if a["completed"] else "‚ùå Not Done"
                        print(f"{i}. {a['task']} (Due: {a['due']}) - {status}")

            elif choice == "3":
                if not assignments:
                    print("You have no assignments to mark complete.")
                else:
                    for i, a in enumerate(assignments, start=1):
                        print(f"{i}. {a['task']}")
                    num = int(input("Enter the number of the assignment to mark complete: "))
                    if 1 <= num <= len(assignments):
                        assignments[num - 1]["completed"] = True
                        print(f"üéâ '{assignments[num - 1]['task']}' marked as complete!")
                    else:
                        print("Invalid selection.")

            elif choice == "4":
                if not assignments:
                    print("You have no assignments to delete.")
                else:
                    for i, a in enumerate(assignments, start=1):
                        print(f"{i}. {a['task']}")
                    num = int(input("Enter the number of the assignment to delete: "))
                    if 1 <= num <= len(assignments):
                        removed = assignments.pop(num - 1)
                        print(f"üóëÔ∏è '{removed['task']}' has been deleted.")
                    else:
                        print("Invalid selection.")

            elif choice == "5":
                print("Returning to Main Menu...")
                break

            else:
                print("Invalid option. Please try again.")
    
#Grade Calculator (need to add decimal capability)
    if function==2:
    ###Inform user what option they have chosen
        print("")
        print("You have chosen to calculate your final grade")
        print("")
    ###Prompt user for category amounts so we ask for the proper amount of inputs
        categories = int(input("Please enter the total assignment categories that you have"))
    ###creating an empty variables so we have a place to store the values as the loop goes
        names = []
        weights = []
        grades = []
        final_grade=0
        for x in range(1,categories+1):
            print(x)
            name = input("Category Name:")
            weight = float(input("Category Weight (as a decimal):"))
            grade = float(input("Final Grade for Category:"))
        ###adding to each empty list
            names.append(name)
            weights.append(weight)
            grades.append(grade)

            calculate = weight*grade
            final_grade += calculate
        print("")
        print("Your final grade is:")
        print(final_grade)
    ###add loop to make continous

    
#Daily Affirmations (completed)
    elif function==3:
        print("")
        print("_____________________________________________________________")
        print("Here's some kind words to lift your spirits:")
        print("")


        affirmations = [
            "    Aren't you an adult?! We don't have time for hand holding!!",
            "    Mistakes are proof that you are trying, don't give up",
            "    Take a deep breath...this situation is only temporary",
            "    It always seems impossible until it's done",
            "    Your effort is what's important, not the grade at the end"]
        import random as rand #renaming just because I can

        print(rand.choice(affirmations))


#Exit    
    elif function==4:
        print("Goodbye")
        break
else:
    print("")
    
```
