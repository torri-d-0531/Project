```python
# Weather-Based Outfit Generator

print("Welcome to the Weather-Based Outfit Generator!")
print("----------------------------------------------")

# Get user input
temperature = float(input("Enter the temperature (°F): "))
condition = input("Enter the weather condition (sunny, rainy, snowy, cloudy, windy): ").lower()

# Outfit recommendation logic
outfit = []
accessories = []

# Temperature-based logic
if temperature < 40:
    outfit.append("a heavy winter coat")
    outfit.append("thick pants or jeans")
    accessories.append("gloves and a scarf")
elif 40 <= temperature < 65:
    outfit.append("a hoodie or light jacket")
    outfit.append("jeans or leggings")
elif 65 <= temperature < 80:
    outfit.append("a t-shirt or light long-sleeve")
    outfit.append("comfortable pants, leggings, or shorts")
else:
    outfit.append("a tank top or breathable t-shirt")
    outfit.append("shorts")
    accessories.append("stay hydrated — it’s hot!")

# Weather condition adjustments
if condition == "rainy":
    accessories.append("an umbrella")
    accessories.append("waterproof boots")
elif condition == "snowy":
    accessories.append("winter boots")
    accessories.append("a warm hat")
elif condition == "windy":
    accessories.append("a windbreaker jacket")
elif condition == "sunny":
    accessories.append("sunglasses")

# Display results
print("\n--- Outfit Recommendation ---")

print("You should wear:")
for item in outfit:
    print(f"- {item}")

if accessories:
    print("\nDon't forget:")
    for item in accessories:
        print(f"- {item}")

print("\nStay safe and dress comfortably!")
```
